```python
import pandas as pd
import numpy as np
```

```python
##Load le fichier excel des réponses
data = pd.read_excel("FFQ.xlsx")
data
```

```python
##Isoler les valeurs
answers = data.iloc[1:, 1:]
answers = answers.fillna(0)
answers = np.array(answers)

##Réponses différentes par question

#Trouver la question 9 (questions non-analysées après)
q9 = data.iloc[0].index.tolist().index('Q9') #nom de la colonne

#Questions 1-8
answers = data.iloc[1:, 1:q9]
answers = answers.fillna(0)
answers = np.array(answers)
answers_flat = answers.reshape(-1)
unique_answers1 = set()
for answer in answers_flat:
    unique_answers1.add(answer)
unique_answers1 = list(unique_answers1)
print(unique_answers1)

#Questions 9-23
#unique_answers2 = []
#for index in range(q9, answers.shape[1]):
    #unique_per_question = set()
    #for answer in answers[:,index]:
        #unique_per_question.add(answer)
    #unique_answers2.append(list(unique_per_question))
#print(unique_answers2)
```

```python
#Valeurs numériques attribuées aux réponses : portions par semaine
dict_portions = { 0 : 0,
            '6x par Jour': 6*7, 
            '5-6x par Sem.' : 5.5, 
            '1x par Sem.' : 1, 
            '2-4x par Sem.' : 3,
            'Jamais ou <1/Mois' : 0,
            '2-3x par Jour' : 2.5*7,
            '4-5x par Jour' : 4.5*7, 
            '1-3x par Mois' : 2/4, 
            '1x par Jour' : 7 }

#Création d'un nouveau tableau avec les valeurs de portions par semaine
portions = []
for index in range(answers.shape[0]):
    portions_per_question = []
    for answer in answers[index,:]:
        portions_per_question.append(dict_portions[answer])
    portions.append(portions_per_question)
portions = np.array(portions)
portions #Lignes = personnes, Colonnes = questions
```

```python
#Load le fichier excel des taux
data_Glu = pd.read_excel('taux.xlsx', sheet_name = "AFTER CONV")
data_Glu = data_Glu["Taux d'acide glutama [mg]"]
data_Glu
```

```python
Glu = np.array(data_Glu)
Glu = Glu.reshape(-1)

#Quantité totale de glutamate par semaine par question en ug
total_per_question = portions * Glu

#Quantité totale de glutamate par semaine en mg
total_per_week = total_per_question.sum(axis=1)

total_per_week
```

```python
cid_to_Glu = {'CID': data.CID[1:], 'Glutamate par semaine [ug]': total_per_week}
df_cid_to_Glu = pd.DataFrame(data = cid_to_Glu).set_index('CID')
df_cid_to_Glu
```

```python
#Convert file to excel
df_cid_to_Glu.to_excel('Glutamate_scoring.xlsx')
```
